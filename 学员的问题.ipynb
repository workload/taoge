# 如何反推仿射变换矩阵

```python
# https://shapely.readthedocs.io/en/latest/manual.html#affine-transformations
```

```python
from shapely import affinity
import numpy as np
from shapely.geometry import Point,LineString,LinearRing,Polygon
```

```python
triangle = Polygon([(1,1),(2,3),(3,1)])
triangle
```

```python
matrix = np.array([0,1,3,0,0,0])  # [a, b, d, e, xoff, yoff]
```

```python
trianglex = affinity.affine_transform(triangle,matrix)
trianglex
```

```python
trianglex.exterior.coords[:]
```

```python
org_coords = triangle.exterior.coords[:3]
des_coords = trianglex.exterior.coords[:3]
```

```python
org_coords,des_coords
```

```python
org = np.mat(org_coords).T
dst = np.mat(des_coords).T
```

```python
org,dst
```

```python
org = np.vstack([org,np.mat([[1,1,1]])])
dst = np.vstack([dst,np.mat([[1,1,1]])])
```

```python
org,dst
```

```python
dst * np.linalg.inv(org) 
```
